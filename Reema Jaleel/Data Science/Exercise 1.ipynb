```python
#Simple Calculator

def add(a,b):
    print("\nAddition => ",a," + ",b," = ",a+b)
    
def sub(a,b):
    print("\nSubstration => ",a," - ",b," = ",a-b)
    
def mult(a,b):
    print("\nMultiplication => ",a," * ",b," = ",a*b)
    
def div(a,b):
    if (b==0):
        print("\nDivision by zero is not possible")
    else:
        print("\nDivision => ",a," / ",b," = ",a/b)
print("Simple Calulator")
print("----------------")
    
a = int(input("Enter a number : "))
b = int(input("Enter another number : "))

while True:
    
    print("\n1.Addition\n2.Substration\n3.Multiplication\n4.Division")
    
    c = int(input("\nEnter your choice : \n"))
    
    if (c==1):
        add(a,b)
    elif (c==2):
        sub(a,b)
    elif (c==3):
        mult(a,b)
    elif (c==4):
        div(a,b)
    else:
        print("Exiting")
        break
```
